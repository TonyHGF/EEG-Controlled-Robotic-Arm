```python
import numpy as np
from matplotlib import pyplot as plt

```

```python
# Load EEG data
eeg_data = np.load("../data/1721289003.8152642/eeg_data_0.npy")

print(eeg_data.shape)

for i in range(8):
    print(eeg_data[:, i].shape)
    plt.subplot(4, 2, i + 1)
    plt.plot(eeg_data[:, i])
    plt.title(f"Channel {i + 1}")

plt.tight_layout()
plt.show()
    
```
