```python
import numpy as np
from matplotlib import pyplot as plt

```

```python
# Load EEG data
eeg_data = np.load("../data/eeg_data.npy")

print(eeg_data.shape)

for i in range(8):
    plt.subplot(4, 2, i + 1)
    plt.plot(eeg_data[i])
    plt.title(f"Channel {i + 1}")

plt.tight_layout()
plt.show()
    
```
